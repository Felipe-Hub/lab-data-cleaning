## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
weather = pd.read_csv('../weather-raw.csv')
weather.head()
```

```python
w_null = weather.isnull().sum()
weather_todrop = list(w_null[w_null == 22].index)
print(weather_todrop)
# dropped the days that had no values (22 NaN), and other useless info
dropped = weather.drop(['id', 'year'], axis=1).set_index('month')
dropped2 = dropped.drop(weather_todrop, axis=1)
clean = dropped2.T

# list_to_drop = stats[(stats['count'] == 0)]
print(clean)
```
